```python
import pandas as pd
import requests
import json
```

```python
database_id = '6c8a44d0c78f402b9a30cdedaa5ee3b1'
token = ''

url = f"https://api.notion.com/v1/databases/{database_id}/query"

payload = {
    #"filter": "string",
    #"start_cursor": "string",
    "page_size": 100
}
headers = {
    "accept": "application/json",
    "Notion-Version": "2022-06-28",
    "content-type": "application/json",
    'Authorization': f'Bearer {token}'
}

response = requests.post(url, json=payload, headers=headers)
```

```python

def transform_data(json_string):
    data = json.loads(json_string)
    properties = data.get('properties', {})
    transformed_data = {}

    for key, value in properties.items():
        if 'plain_text' in value.get('rich_text', [{}])[0]:
            transformed_data[key] = value['rich_text'][0]['plain_text']

    return transformed_data
```

```python
transform_data(str(data['results'][-1]))
```

```python

# Defina o token de integração da API do Notion
#token = 'seu_token_de_integracao'

# Defina o ID da página (linha) que deseja atualizar
page_id = '4358c39a-c986-425b-85fe-626e25622dd0'

# Defina os dados para atualizar a página
data = {
    'properties': {
        'Pode compartilhar exemplo dos dados?': {
            'rich_text': [{'text': {'content': 'Novo valor do nome'}}]
        },
        'A empresa já realiza a atividade hoje? Se sim, como é feito?': {
            'rich_text': [{'text': {'content': 'Nova descrição'}}]
        }
    }
}

# Faça a solicitação PATCH para atualizar a página
url = f'https://api.notion.com/v1/pages/{page_id}'
headers = {
    'Authorization': f'Bearer {token}',
    'Content-Type': 'application/json',
    'Notion-Version': '2021-05-13'
}
response = requests.patch(url, json=data, headers=headers)
```
